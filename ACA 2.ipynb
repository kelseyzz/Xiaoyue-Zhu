Feature selection

1. Drop ID-related features

ID is unique, where the derivation is significantly large and samples share no common infomation in this feature, thus does not benefit the model.

2. Drop Transitions/Durations

where the derivation is close to 0 and samples virtually reveal no difference in this feature, thus does not benefit the model.

3. Drop non-instructional strategies

which seems not to contribute to the classification

```python
def drop_feature(data):
    data.drop(['UNIQUEID', 'STUDENTID'], axis=1, inplace=True)
    data.drop(['Transitions/Durations'], axis=1, inplace=True)
    data.drop(['Class','SCHOOL'], axis=1, inplace=True)
```

Discretization for numerical features & Chi-square test for finally selecting the best k features

```python
from sklearn.preprocessing import KBinsDiscretizer
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
def discretize(data):
    numeric = data[data.describe().columns]
    numeric_ = numeric.drop('Gender', axis=1)  # remove Gender
    numeric_ = KBinsDiscretizer(n_bins=3, encode='onehot', strategy='quantile').fit_transform(numeric_)
    numeric = np.concatenate((numeric['Gender'].values.reshape(-1, 1), numeric_.toarray()), axis=1)  # Gender + numeric

    categorical = data[data.describe(include='O').columns]
    categorical = pd.get_dummies(categorical, drop_first=True)

    feature = np.concatenate((numeric, categorical.drop('ONTASK_Y', axis=1)),axis=1)
    label = categorical['ONTASK_Y']
  

    feature = SelectKBest(chi2, k=10).fit_transform(feature,label)
    
    return (feature, label)
```

```python
def feature_selection(data):
    drop_feature(data)
    return discretize(data)
```
